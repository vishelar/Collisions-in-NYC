```python
import pandas as pd
import numpy as np
import pysal
import geopandas as gp
import matplotlib.pyplot as pl
%matplotlib inline
```

```python
collisions = pd.read_csv('nyc_collision_2015.csv')
```

```python
collisions_all = pd.read_csv('NYC_collision.csv')
```

```python
collisions.drop('Unnamed: 0', axis= 1, inplace= True)
```

```python
collisions_all.drop('Unnamed: 0', axis= 1, inplace= True)
```

```python
collisions['date'] = pd.to_datetime(collisions.date)
```

```python
collisions_all['date'] = pd.to_datetime(collisions_all.date)
```

```python
collisions_all.date.describe()
```

```python
collisions.date.describe()
```

```python
coll_alcohol = collisions_all[collisions_all.contributing_factor == 'Alcohol Involvement']
coll_alcohol.reset_index(inplace= True)
coll_alcohol.drop(['index'], axis=1, inplace=True)
```

```python
locs = zip(coll_alcohol.loc[:,'long'], coll_alcohol.loc[:,'lat'])
```

```python
coll_alcohol.loc[:,'location'] = pd.Series(zip(coll_alcohol.loc[:,'long'], coll_alcohol.loc[:,'lat']))
```

```python
collisions_all.columns
```

```python
collisions_all.loc[:,'injuries'] = collisions_all.loc[:,'pedestrian_injured'] + collisions_all.loc[:,'cyclist_injured'] + collisions_all.loc[:,'motorist_injured']
```

# Higest Injuries

```python
coll_top = collisions_all.groupby(['lat','long']).agg({'injuries':'sum','pedestrian_injured':'sum','motorist_injured':'sum', 'cyclist_injured':'sum'  }).sort_values('injuries',ascending = False)[0:5].reset_index()
coll_top_ped = collisions_all.groupby(['lat','long']).agg({'injuries':'sum','pedestrian_injured':'sum','motorist_injured':'sum', 'cyclist_injured':'sum'  }).sort_values('pedestrian_injured',ascending = False)[0:5].reset_index()
coll_top_cyc = collisions_all.groupby(['lat','long']).agg({'injuries':'sum','pedestrian_injured':'sum','motorist_injured':'sum', 'cyclist_injured':'sum'  }).sort_values('cyclist_injured', ascending = False)[0:5].reset_index()
coll_top_mot = collisions_all.groupby(['lat','long']).agg({'injuries':'sum','pedestrian_injured':'sum','motorist_injured':'sum', 'cyclist_injured':'sum'  }).sort_values('motorist_injured',ascending = False)[0:5].reset_index()
```

```python
coll_top = collisions_all.groupby(['lat','long']).agg({'injuries':'sum','pedestrian_injured':'sum','motorist_injured':'sum', 'cyclist_injured':'sum'  }).sort_values('injuries',ascending = False)[0:5].reset_index()
```

```python
coll_top_ped.to_csv('coll_top_ped.csv')
coll_top_cyc.to_csv('coll_top_cyc.csv')
coll_top_mot.to_csv('coll_top_mot.csv')
coll_top.to_csv('coll_top.csv')
```

# Season

```python
for i in xrange(len(collisions)):
    collisions.loc[i,'month'] = collisions.date[i].month
```

```python
collisions_summer = collisions[((collisions.month)>= 6 ) & 
                            ((collisions.month) <= 9)]
# Split data into winter months
collisions_winter = collisions[((collisions.month) >= 12 ) | 
                            ((collisions.month) <= 3)]
```

```python
collisions_summer.to_csv('collisions_summer.csv')
collisions_winter.to_csv('collisions_winter.csv')
```
